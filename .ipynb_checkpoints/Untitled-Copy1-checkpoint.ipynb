```python
# ================================
# 1. 导入必要库（Import Libraries）
# ================================
import numpy as np
import pandas as pd

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
print(1)
```

```python
# ================================
# 2. 数据加载与预处理（Data Loading）
# ================================
df = pd.read_csv('energy_dataset.csv')
df['time'] = pd.to_datetime(df['time'], utc=True)
df = df.sort_values('time').reset_index(drop=True)

target = 'total load actual'
df = df[['time', target] + [col for col in df.columns if col not in ['time', target]]]
df = df.dropna(axis=1, how='all')           # 删除全空列
df[target] = df[target].interpolate()       # 目标插值

print(f"数据形状: {df.shape} | 时间范围: {df['time'].min()} → {df['time'].max()}")

```

```python
# ================================
# 3. 特征工程（Feature Engineering）
# ================================
def create_features(df):
    df = df.copy()
    df['hour'] = df['time'].dt.hour
    df['dayofweek'] = df['time'].dt.dayofweek
    df['month'] = df['time'].dt.month
    df['is_weekend'] = (df['dayofweek'] >= 5).astype(int)

    # 滞后特征
    df['load_lag_24'] = df[target].shift(24)
    df['load_lag_48'] = df[target].shift(48)
    df['load_lag_168'] = df[target].shift(168)  # 一周前

    # 滚动统计
    df['load_roll_mean_24'] = df[target].shift(24).rolling(24).mean()
    df['load_roll_std_24'] = df[target].shift(24).rolling(24).std()

    return df
    
df_feat = create_features(df)
df_feat = df_feat.dropna().reset_index(drop=True)
```

```python
print(type(df_feat))
```

```python
# ================================
# 4. 时序划分（Temporal Split）
# ================================
n = len(df_feat)
train_size = int(n * 0.70)
val_size = int(n * 0.15)

train_df = df_feat[:train_size]
val_df = df_feat[train_size:train_size + val_size]
test_df = df_feat[train_size + val_size:]

print(f"训练集: {len(train_df)} | 验证集: {len(val_df)} | 测试集: {len(test_df)}")
```

```python
# ================================
# 5. 标准化 & 数据集（Scaling & Dataset）
# ================================
feature_columns = [c for c in df_feat.columns if c not in [target, 'time']]

scaler_X = StandardScaler().fit(train_df[feature_columns])
scaler_y = StandardScaler().fit(train_df[[target]])


class LoadDataset(Dataset):
    def __init__(self, df, seq_len=168):
        self.seq_len = seq_len
        self.X = scaler_X.transform(df[feature_columns])
        self.y = scaler_y.transform(df[[target]]).reshape(-1)

    def __len__(self):
        return len(self.X) - self.seq_len

    def __getitem__(self, idx):
        return (torch.FloatTensor(self.X[idx:idx + self.seq_len]),
                torch.FloatTensor([self.y[idx + self.seq_len]]))


seq_len = 168
batch_size = 64

train_loader = DataLoader(LoadDataset(train_df, seq_len), batch_size=batch_size, shuffle=True)
val_loader = DataLoader(LoadDataset(val_df, seq_len), batch_size=batch_size, shuffle=False)
test_loader = DataLoader(LoadDataset(test_df, seq_len), batch_size=batch_size, shuffle=False)
```

```python

# ================================
# 6. 三模型定义（Three Models）
# ================================
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(device)

class RNN(nn.Module):
    def __init__(self, input_size, hidden_size=64, num_layers=2):
        super().__init__()
        self.rnn = nn.RNN(input_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, 1)

    def forward(self, x):
        out, _ = self.rnn(x)
        return self.fc(out[:, -1, :])


class LSTM(nn.Module):
    def __init__(self, input_size, hidden_size=128, num_layers=2, dropout=0.2):
        super().__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers,
                            batch_first=True, dropout=dropout)
        self.fc = nn.Linear(hidden_size, 1)

    def forward(self, x):
        out, _ = self.lstm(x)
        return self.fc(out[:, -1, :])


class LSTM_Bi(nn.Module):
    def __init__(self, input_size, hidden_size=128, num_layers=2, dropout=0.2):
        super().__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers,
                            batch_first=True, dropout=dropout, bidirectional=True)
        self.fc = nn.Linear(hidden_size * 2, 1)  # 双向 → 维度翻倍

    def forward(self, x):
        out, _ = self.lstm(x)
        return self.fc(out[:, -1, :])


input_size = len(feature_columns)
models = {
    'RNN': RNN(input_size).to(device),
    'LSTM': LSTM(input_size).to(device),
    'LSTM-Bi': LSTM_Bi(input_size).to(device)
}

```

```python

# ================================
# 7. 训练函数（Training Loop）
# ================================
def train_model(model, name, epochs=80, lr=0.001):
    criterion = nn.MSELoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)
    best_val = float('inf')
    train_losses, val_losses = [], []

    print(f"\n正在训练 → {name} (epochs={epochs})")
    for epoch in range(epochs):
        model.train()
        train_loss = 0
        for x, y in train_loader:
            x, y = x.to(device), y.to(device)
            optimizer.zero_grad()
            pred = model(x)
            loss = criterion(pred, y)
            loss.backward()
            optimizer.step()
            train_loss += loss.item()

        # Validation
        model.eval()
        val_loss = 0
        with torch.no_grad():
            for x, y in val_loader:
                x, y = x.to(device), y.to(device)
                pred = model(x)
                val_loss += criterion(pred, y).item()

        train_loss /= len(train_loader)
        val_loss /= len(val_loader)
        train_losses.append(train_loss)
        val_losses.append(val_loss)

        if val_loss < best_val:
            best_val = val_loss
            torch.save(model.state_dict(), f'best_{name}.pth')

        if (epoch + 1) % 20 == 0:
            print(f"Epoch {epoch + 1:2d} | Train Loss: {train_loss:.5f} | Val Loss: {val_loss:.5f}")

    return train_losses, val_losses

```

```python
# ================================
# 8. 训练三个模型
# =============================1
history = {}
for name, model in models.items():
    train_l, val_l = train_model(model, name, epochs=80, lr=0.001)
    history[name] = {'train': train_l, 'val': val_l}
```

```python
# ================================
# 9. 测试与可视化（Evaluation）
# ================================
results = []


def evaluate(model, name):
    model.load_state_dict(torch.load(f'best_{name}.pth'))
    model.eval()
    preds, actuals = [], []
    with torch.no_grad():
        for x, y in test_loader:
            pred = model(x.to(device)).cpu().numpy()
            preds.extend(pred.flatten())
            actuals.extend(y.numpy().flatten())

    # 反标准化
    preds = scaler_y.inverse_transform(np.array(preds).reshape(-1, 1)).flatten()
    actuals = scaler_y.inverse_transform(np.array(actuals).reshape(-1, 1)).flatten()

    mae = mean_absolute_error(actuals, preds)
    rmse = np.sqrt(mean_squared_error(actuals, preds))
    r2 = r2_score(actuals, preds)

    results.append({'Model': name, 'MAE (MW)': round(mae, 2), 'RMSE (MW)': round(rmse, 2), 'R²': round(r2, 4)})

    print(f"\n{name} → MAE: {mae:.2f} MW | RMSE: {rmse:.2f} MW | R²: {r2:.4f}")

    # 预测曲线（最后500点）
    plt.figure(figsize=(15, 5))
    plt.plot(actuals[-500:], label='true load demand', linewidth=2)
    plt.plot(preds[-500:], label='predicted load demand', linewidth=2, alpha=0.9)
    plt.title(f'{name} - Electricity demand forecast (Last 500 hours of the test set)', fontsize=16)
    plt.xlabel('time')
    plt.ylabel('load demand (MW)')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()


for name in models:
    evaluate(models[name], name)
```

```python

# ================================
# 10. 最终对比图表（Final Comparison）
# ================================
# 性能表格
print("\n" + "=" * 70)
print("                   三模型最终性能对比")
print("=" * 70)
results_df = pd.DataFrame(results)
print(results_df.to_string(index=False))

# 损失曲线对比
plt.figure(figsize=(12, 6))
for name in history:
    plt.plot(history[name]['train'], label=f'{name} Train')
    plt.plot(history[name]['val'], label=f'{name} Val', linestyle='--')
plt.title('三模型训练与验证损失曲线对比')
plt.xlabel('Epoch')
plt.ylabel('MSE Loss')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
